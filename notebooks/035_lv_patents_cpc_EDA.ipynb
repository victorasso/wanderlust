## CPC Sections

The CPC has 9 sections defined as the following

https://www.epo.org/searching-for-patents/helpful-resources/first-time-here/classification/cpc.html

```python
CPC_sections = {
    'A': 'Human necessities',
    'B': 'Performing operations; transporting',
    'C': 'Chemistry; metallurgy',
    'D': 'Textiles; paper',
    'E': 'Fixed constructions',
    'F': 'Mechanical engineering; lighting; heating; weapons; blasting engines or pumps',
    'G': 'Physics',
    'H': 'Electricity',
    'Y': 'General' # General tagging of new technological developments; general tagging of cross-sectional technologies spanning over several sections of the IPC; technical subjects covered by former USPC cross-reference art collections [XRACs] and digests
}
```

```python
def get_cpc_sections(s):
  return s[0]

df['CPC_section'] = df['cpc_classifications'].apply(get_cpc_sections)
pd.DataFrame(df.groupby('CPC_section').size(), columns=['size'])
```

# CPC definitions

The official table of contents for CPC definitions are here

https://www.cooperativepatentclassification.org/cpcSchemeAndDefinitions/table

```python
def get_cpc_definitions(s):
  return ''.join(list(s)
  [:4])

df['CPC_definition'] = df['cpc_classifications'].apply(get_cpc_definitions)
pd.DataFrame(df.groupby('CPC_definition').size(), columns=['size'])
```
